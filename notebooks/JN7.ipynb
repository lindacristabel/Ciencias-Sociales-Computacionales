```python
import wikipediaapi
import pandas as pd
pd.set_option("display.notebook_repr_html", False)
from itertools import islice
```

```python
wiki_wiki = wikipediaapi.Wikipedia('Ciencias Sociales Computacionales (can06728@uvg.edu.gt)', 'en')
page_py = wiki_wiki.page('Guatemala')
```

```python
print("Página existe: %s" % page_py.exists())
print("Página - Título: %s" % page_py.title)
print("Página - Secciones: %s" % page_py.sections)
```

```python
cat_Historia = wiki_wiki.page("Category:History")
print("page exist: %s" % cat_Historia.exists())
```

```python
def print_categorymembers(categorymembers, level=0, max_level=1):
    for c in categorymembers.values():
        print("%s: %s (ns: %d)" % ("*" * (level +1), c.title, c.ns))
        if c.ns == wikipediaapi.Namespace.CATEGORY and level < max_level:
            print_categorymembers(c.categorymembers, level=level + 1, max_level=max_level)

print("Miembros de la categoría: Category:History")
print_categorymembers(cat_Historia.categorymembers)
```

```python
pages_names = cat_Historia.categorymembers
print(pages_names)
```

```python
data = []
for key, value in islice (pages_names.items(), 1, None):
    page_py = wiki_wiki.page(key)
    data.append(
        {
            'Title': page_py.title,
            'Summary': page_py.summary,
            'Content': page_py.sections
        }
    )
df = pd.DataFrame(data)
```
