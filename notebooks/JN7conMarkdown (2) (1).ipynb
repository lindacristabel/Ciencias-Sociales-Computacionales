En las siguientes líneas de texto creamos una nueva variable, pedimos que busque la página que le pedimos y que nos diga si existe o no existe:

```python
cat_Historia = wiki_wiki.page("Category:History")
print("page exist: %s" % cat_Historia.exists())
```

En las siguientes líneas vamos a hacer una búsqueda de todo lo que existe en esta categoría:
Vamos a definir una nueva función, donde vamos a buscar la información que contiene la categoría que nos interesa 
Luego vamos a pedir que imprima los datos que solicitamos (el titulo de cada categoria y el nivel de cada una)

```python
def print_categorymembers(categorymembers, level=0, max_level=1):
    for c in categorymembers.values():
        print("%s: %s (ns: %d)" % ("*" * (level +1), c.title, c.ns))
        if c.ns == wikipediaapi.Namespace.CATEGORY and level < max_level:
            print_categorymembers(c.categorymembers, level=level + 1, max_level=max_level)

print("Miembros de la categoría: Category:History")
print_categorymembers(cat_Historia.categorymembers)
```

En la siguiente línea de código pedimos que nos de los nombres de las páginas que vamos a buscar
En la nueva variable vamos a guardar las categorías que creamos
pedimos que nos imprima todas las variables:

```python
pages_names = cat_Historia.categorymembers
print(pages_names)
```

Vamos a crear una nueva variable, un listado de toda la información que tenemos
Vamos a guardar la información en esta nueva variable
Vamos a dividir la información en (titulo, resumen y contenido)
Pedimos que nos haga una base de datos con la información

```python
data = []
for key, value in islice (pages_names.items(), 1, None):
    page_py = wiki_wiki.page(key)
    data.append(
        {
            'Title': page_py.title,
            'Summary': page_py.summary,
            'Content': page_py.sections
        }
    )
df = pd.DataFrame(data)
```

En la siguiente línea de código pedimos que despliegue la base de datos que creamos:

```python
display(df)
```

En la siguiente línea de código guardamos la base de datos que creamos como un archivo csv

```python
df.to_csv('../data/base_datos_wikipedia_Historia.csv')
```
